```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

df=pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
df.shape
```

```python
%matplotlib inline
plt.title("prediction income of Canada")
plt.xlabel("year in AD")
plt.ylabel("Income US $")
plt.plot(df.year,df.income)
plt.show()


```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.income)

```

```python
reg.predict([[1990]])
```

```python
#calculating the slope i.e value of  m
reg.coef_
```

```python
#calculating the intercept i.e value of b
reg.intercept_
```

```python
#cacualte the income of 1990
#usinf y=mx+b
828.46507522*1990+(-1632210.7578554575)
```
